```python
import os
import csv
import glob
from pathlib import Path
from time import gmtime, strftime
```

```python
#Counting lines in CSV files to see which is the shortest and which is the longest
file_counter = 0
shortest = 0
longest = 0
record_path = Path("Records")
csv_files = record_path.glob('*.csv')
for filename in csv_files:
    
    with open(filename) as f:
        nr_lines = sum(1 for line in f)
   
    print(f'{filename.__str__()} -> {nr_lines}')

    if file_counter == 0:
        shortest = nr_lines
        longest = nr_lines
    
    if nr_lines < shortest:
        shortest = nr_lines
        
    if nr_lines > longest:
        longest = nr_lines
            
    file_counter+=1
    
print(f'Shortest: {shortest}, Longest: {longest}')
```

```python
record_summary_time = strftime('%Y-%m-%d-%H-%M-%S', gmtime());
record_summary_folder = 'RecordSummary'
summary_csv_file_name = f'{record_summary_folder}\\{record_summary_time}.csv'

if not os.path.exists(record_summary_folder):
    os.makedirs(record_summary_folder)

file_counter = 0;
record_path = Path("Records")
csv_files = list(filter(os.path.isfile,record_path.glob('*.csv')))
csv_files.sort(key=lambda x: os.path.getmtime(x))

with open(summary_csv_file_name, mode='w', newline='') as summary_file:
    summary_writer = csv.writer(summary_file)
    
    for filename in csv_files:
        filename_parts = filename.__str__().replace('Records\\','').split("_")
        eeg_object_name = filename_parts[3] #up or down
        
        with open(filename) as csv_file:
            data = list(csv.reader(csv_file))
            tp = list(map(list, zip(*data)))
            concatenated_list = [eeg_object_name] + tp[1][1:shortest] + tp[2][1:shortest] + tp[3][1:shortest] + tp[4][1:shortest]
            summary_writer.writerow(concatenated_list)

        file_counter+=1;
    
    print(f'Merged {file_counter} files to {summary_csv_file_name}')    
```
